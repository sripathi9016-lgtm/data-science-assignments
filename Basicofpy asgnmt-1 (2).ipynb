```python
###excercise-01
def is_prime(n):
    
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example check
num = 30
print(f"Is {num} prime?", is_prime(num))
         
```

```python
# Hospital Billing System Calculation

# Given values
room_charge_per_day = 2000
doctor_consultation_fee = 1500
lab_test_charge = 300
days_stayed = 4
num_lab_tests = 3
medicine_charges = 2400

# Calculate individual components
room_total = room_charge_per_day * days_stayed
lab_total = lab_test_charge * num_lab_tests

# Calculate total amount before discount
total_bill_before_discount = room_total + doctor_consultation_fee + lab_total + medicine_charges

# Apply discount if applicable
if total_bill_before_discount > 10000:
    discount = 0.10 * total_bill_before_discount
else:
    discount = 0

final_bill = total_bill_before_discount - discount

# Print detailed bill
print("Hospital Bill Details:")
print(f"Room charges for {days_stayed} days: ₹{room_total}")
print(f"Doctor consultation fee: ₹{doctor_consultation_fee}")
print(f"Lab test charges for {num_lab_tests} tests: ₹{lab_total}")
print(f"Medicine charges: ₹{medicine_charges}")
print(f"Total bill before discount: ₹{total_bill_before_discount}")
print(f"Discount applied: ₹{discount}")
print(f"Final bill to be paid: ₹{final_bill}")
```
