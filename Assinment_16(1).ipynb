```python
#tfidf vectorzer to convert genre to numerical features for cosine similaririties
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidfs=TfidfVectorizer(stop_words="english")
```

```python
scores=tfidfs.fit_transform(df["genre"])
```

```python
#find the cosine similarity
similarity_scores=cosine_similarity(scores)
```

```python
similarity_scores
```

```python
#function to find the recomendation sytems
def animes_recomendation(titles,threshold):
    if titles not in df["name"].values:
        print("given animes name is not found ")
        return
    animes_index=df[df['name'] == titles].index[0]
    animes_similar=pd.Series(similarity_scores[animes_index],index=df.index)
    animes_similar=animes_similar[(animes_similar>=threshold)&(animes_similar!=animes_index)]
    animes_similar=animes_similar.sort_values(ascending=False)
    top_animes=animes_similar.head(20)
    print(f"the top {len(top_animes)} animes are similar to {titles} with threshold>={threshold}")
    print(df.loc[top_animes.index,["name","genre","rating"]])
    return top_animes
```

```python
animes_recomendation("Sci-Fi, Thriller",0.5)
```

```python
animes_recomendation("Naruto",0.5)
```

```python
animes_recomendation("Naruto",0.9)
```

User-Based vs Item-Based Collaborative Filtering
User-Based

Finds similar users.

Recommends items liked by those users.

Item-Based

Finds similar items.

Recommends items similar to what the user already liked.

In short:
User-based = people like you

Item-based = items like this

What is Collaborative Filtering?

Collaborative filtering makes recommendations by using patterns in user behavior.
If users like similar things, the system predicts what you may like based on others with similar tastes.

Simple idea:

People with similar choices help predict your choices.
